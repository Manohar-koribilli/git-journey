```python
a = 95
b = 6

quotient = a // b
remainder = a % b

result = [b] * quotient

if remainder != 0:
    result.append(remainder)
    print(result, "=", quotient, "+ 1")
else:
    print(result, "=", quotient)
```

```python
a =95
b = 6

quotient = a // b
remainder = a % b

result = [b] * quotient

if remainder != 0:
    result.append(remainder)
    print(f"{result} = {b}*{quotient} + {remainder}*1 = {a}")
else:
    print(f"{result} = {b}*{quotient} = {a}")
```
